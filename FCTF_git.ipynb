```python
data = pd.read_csv('/content/test2/GBM_GBS_Data.csv')
augmented_data = pd.concat([data] *2, ignore_index=True)
X = augmented_data.drop(['Y'], axis=1)
y = augmented_data['Y']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
Train = pd.concat([X_train, y_train], axis=1)
Test = pd.concat([X_test, y_test], axis=1)
__builtins__.X_train = X_train
__builtins__.X_test = X_test
__builtins__.Train = Train
__builtins__.Test = Test
```

```python
# trainer and test_loaders
import torch
from torch.utils.data import DataLoader, Dataset
import pandas as pd
import numpy as np

class MyDataset(Dataset):
    def __init__(self, data, targets):
        self.data = data
        self.targets = targets

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        x = torch.tensor(self.data[idx], dtype=torch.floatest2)
        y = torch.tensor(self.targets[idx], dtype=torch.long)
        return x, y

batch_size = 32

train_dataset = MyDataset('/content/test2/train_embedding_fold.pth', '/content/test2/test_embedding_fold.pth')
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)

test_dataset = MyDataset('/content/test2/train_embedding_fold.pth', '/content/drive/test2/test_embedding_fold.pth')
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=True)

__builtins__.test_loader = test_loader
__builtins__.train_loader = train_loader
```
